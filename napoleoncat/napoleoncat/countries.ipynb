```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options  
from bs4 import BeautifulSoup
import requests
```

```python
chrome_options = Options()  
chrome_options.add_argument("--headless")
driver = webdriver.Chrome(options=chrome_options)
driver.get("https://napoleoncat.com/stats")
```

```python
# get input field
input_field = driver.find_element_by_class_name('select2-search__field')
# click input field to show countries
input_field.click()
# get countries
countries = driver.find_elements_by_class_name('select2-results__option')
```

```python
df_countries = pd.DataFrame(columns=['country_name', 'url'])
# get countries names and urls
for country in countries:
    country_name = country.text
    url = "https://"+'/'.join(country.get_attribute('id').split('https://')[1].split('/')[:-2])
    df_countries = df_countries.append({
        'country_name': country_name,
        'url': url
    }, ignore_index=True)
```

```python
df_result = pd.DataFrame(columns = list(df_countries.columns))
```
